```python
#crosswalk_df = pd.read_csv('/Users/shailjasomani/Documents/USD_MS_ADS/ADS_599/Datasets/fips_zip_x.txt', 
#                           delimiter=',', quotechar='"', skiprows=1, encoding='ISO-8859-1')

crosswalk_df = pd.read_csv('/Users/roger.qiu/Documents/other/code/USD/12. Capstone Project/data/fips_zip_x.txt', 
                           delimiter=',', quotechar='"', skiprows=1, encoding='ISO-8859-1')

crosswalk_df.head()
```

```python
# Keep only crosswalk columns needed
county_to_ZCTA = crosswalk_df[['county', 'ZIP Census Tabulation Area']]
county_to_ZCTA.head()
```

### Read in CDC PLACES health measures data (to be joined using ZCTA)
Note: Do not confuse LocationName with the FIPS codes. These are now ZCTAs. One way to check/confirm this is that this LocationName field has 1002, which is a ZCTA code, but not a valid FIPS code.

```python
# Read in CDC PLACES health measures data
health_meas = pd.read_csv('/Users/roger.qiu/Documents/other/code/USD/12. Capstone Project/data/PLACES__Local_Data_for_Better_Health__ZCTA_Data_2023_release_20240629.csv')

#health_meas = pd.read_csv('/Users/shailjasomani/Documents/USD_MS_ADS/ADS_599/Datasets/PLACES__Local_Data_for_Better_Health__ZCTA_Data_2023_release_20240629.csv')
health_meas.head()
```

```python
health_meas.shape
```

```python
# inner join county_ZCTA crosswalk to health measures on county name to get ZCTA
health_meas_with_county = health_meas.merge(county_to_ZCTA, left_on='LocationName', right_on='county')
health_meas_with_county.head()
```

```python
health_meas_with_county.shape
```

```python
# pivot the data so that each row is a county and each field is a metric's % value
health_meas_by_county = health_meas_with_county.pivot_table(index='LocationName', 
                                                            columns='Short_Question_Text', 
                                                            values='Data_Value',
                                                            aggfunc='first') 

# make locationname a column again & remove short_question_text as header
health_meas_by_county.reset_index(inplace=True)
health_meas_by_county.columns.name = None
health_meas_by_county.head()
```

```python
health_meas_by_county.shape
```

```python
target_sdoh_data.head()
```

```python
target_sdoh_data.shape
```

```python
# join health measure by county to target sdoh data by county number to get full dataset
full_data = target_sdoh_data.merge(health_meas_by_county, left_on='COUNTYFIPS', right_on='LocationName')
full_data.head()
```

```python
full_data.shape
```

```python
# write out full_data as a csv to local folder
full_data.to_csv('full_data.csv', index=False)
```
